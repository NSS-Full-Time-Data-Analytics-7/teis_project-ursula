```python
monthly_df = pd.read_excel('../data/TEIS_Referral_Data_for_NSS_Compiled_Monthly.xlsx')
monthly_df.columns = map(str.lower, monthly_df.columns)
monthly_df.columns = monthly_df.columns.str.replace(' ', '_')
monthly_df['exit_date'] = pd.to_datetime(monthly_df['exit_date'], errors = 'coerce')
monthly_df['initial_ifsp_date'] = pd.to_datetime(monthly_df['initial_ifsp_date'], errors='coerce')
monthly_df = monthly_df.drop(columns = ['referral_source_category', '_qtr_', 'qtr', 'late_referral', '2013_child_count', '2012_child_count', 'child_count', 'service_coordinator_counter',  'eligibility_count', 'evaluation_count', 'notification_month', 'tenn_region', 'latest_ifsp_date', 'initial_eligibility', 'initial_eligibility_date', 'parent_consent_date'])
monthly_df['successful'] = np.where(monthly_df['initial_ifsp_date'].isnull(), ("not_successful"), ("successful"))
monthly_df
```

```python
july_df = pd.read_excel('../data/TEIS_Referral_Data_for_Nss_pulled_7-1-22.xlsx')
july_df.columns = map(str.lower, july_df.columns)
july_df.columns = july_df.columns.str.replace(' ', '_')
july_df['successful'] = np.where(july_df['initial_ifsp_date'].isnull(), ("not_successful"), ("successful"))

july_df
```

```python
no_dups_df = monthly_df.loc[monthly_df.duplicated(subset = 'child_id', keep = 'first')]
no_dups_df
```

```python
teis_merge = pd.merge(no_dups_df, july_df, how='outer').sort_values(['child_id'])
teis_merge
```
