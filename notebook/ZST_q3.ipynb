## FIrst: what is the average age of children at time of referral per POE?

```python
# add age column, age calculated in months from DOB to referral date
monthly_df = monthly_df.rename(columns = {'notification/_referral_date':'referral_date'})
monthly_df['birth_year'] = monthly_df['dob'].dt.year
monthly_df['birth_month'] = monthly_df['dob'].dt.month
monthly_df['referral_year'] = monthly_df['referral_date'].dt.year
monthly_df['referral_month'] = monthly_df['referral_date'].dt.month
monthly_df
```

```python
#add age at referral column
monthly_df['age_at_referral_in_months'] = (monthly_df.referral_year - monthly_df.birth_year)*12 + (monthly_df.referral_month - monthly_df.birth_month) 
```

```python
referral_age = monthly_df[['child_id', 'poe', 'dob', 'referral_date', 'age_at_referral_in_months', 'county_ses', 'referral_source_type_name']]
referral_age.sort_values(by = 'child_id')
```

```python
#drop referrals that were referred after 3rd brthday (36 months)
referral_age = referral_age[referral_age.age_at_referral_in_months <= 36]
```

```python
referral_age
```

```python
#subset df for aggregation
age_by_poe = referral_age[['poe', 'age_at_referral_in_months']]
```

### avg age in months at time of referral for each POE

```python

age_by_poe.groupby('poe').mean()
```

### avg age at referral by county_ses

```python
age_by_county_ses = referral_age[['county_ses', 'age_at_referral_in_months']]
age_by_county_ses.groupby('county_ses').mean()
```

### avg age at referral by referral source

```python
age_by_source = referral_age[['referral_source_type_name', 'age_at_referral_in_months']]
age_by_source.groupby('referral_source_type_name').mean()
```

## Count of referrals by POE

```python
referrals_by_poe = monthly_df[['poe', 'referral_count']]
referrals_by_poe.groupby('poe').count()
```

## Count of successful referrals by POE

```python
success_by_poe = monthly_df[['poe', 'initial_ifsp_date']]
success_by_poe.groupby('poe').count()
```

```python
## Percent successful by POE
percent_successful_by_poe = monthly_df[['poe', 'referral_count', 'initial_ifsp_date']]
percent_successful_by_poe = percent_successful_by_poe.groupby('poe').count()
percent_successful_by_poe['percent_successful'] = (percent_successful_by_poe['initial_ifsp_date'] / percent_successful_by_poe['referral_count']) * 100
percent_successful_by_poe
```
