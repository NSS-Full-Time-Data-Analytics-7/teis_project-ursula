what is the breakdown of child actually getting an IFSP by referral source? e.g. do parent referrals lead to the highest chance of getting a Plan? vs. PCP referrals?

```python
denominator = pulled_july['referral_source'].value_counts().reset_index()
denominator = denominator.reset_index()
denominator = denominator.rename(columns = {'level_0':'index','index':'referral_source', 'referral_source':'count'})
denominator
```

```python
successful_only = pulled_july.loc[pulled_july['child_phase'] == 'IFSP']

#value_counts for referral source
#outer merge on index
```

```python
successful_only = successful_only['referral_source'].value_counts().reset_index()
successful_only= successful_only.reset_index()
successful_only = successful_only.rename(columns = {'level_0':'index','index':'referral_source', 'referral_source':'count'})
successful_only
```

```python
merged = successful_only.merge(denominator, how='outer', on='referral_source')
merged['pct_successful'] = merged['count_x'] / merged['count_y']
merged.sort_values(by = 'pct_successful', ascending = False)
```
