```python
pulled_july = pd.read_csv('../../data/Referral_Data_Pulled_7-1-22.csv')
```

```python
pulled_july = pulled_july.rename(columns = {'Child ID':'child_ID','County Name':'county_name', 'County SES':'county_ses', 'Child Phase':'child_phase','Service Coordinator':'service_coord','Notification Date':'notif_date','Parent Consent Date':'consent_date','Referral Source Type Name':'referral_source','Initial Eligibility':'initial_elig','Initial Eligibility Date':'elig_date','Initial IFSP Date':'IFSP_begin','Latest IFSP Date':'IFSP_end','Exit Reason':'exit_reason','Exit Date':'exit_date'})
pulled_july
```

```python
POE_counts = pulled_july.POE.value_counts().to_frame().reset_index()
POE_counts
```

```python
POE_counts['total_referrals'] = 96580
POE_counts
```

```python
POE_counts['POE_makeup'] = (POE_counts['POE'] / POE_counts['total_referrals']).round(decimals=3)*100
POE_counts = POE_counts.rename(columns = {'index':'POE','POE':'count'})
POE_counts
```

```python
successful_referrals = pulled_july.loc[pulled_july['child_phase'] == 'IFSP']
successful_referrals
```

```python
#POE_counts = pulled_july.POE.value_counts().to_frame().reset_index()
POE_successful_ref = successful_referrals.POE.value_counts().to_frame().reset_index()
POE_successful_ref['total'] = 45493
POE_successful_ref['POE_ISFP_pct'] = (POE_successful_ref['POE'] / POE_successful_ref['total']).round(decimals=3)*100
POE_successful_ref = POE_successful_ref.rename(columns = {'index':'POE','POE':'count','total':'IFSP_total'})
POE_successful_ref
```

```python
merged = POE_successful_ref.merge(POE_counts, how='inner', on='POE')
merged['ISFP_conversion_rank'] = merged['POE_ISFP_pct'].rank(ascending=False)
merged['POE_makeup_rank'] = merged['POE_makeup'].rank(ascending=False)
merged[['POE', 'ISFP_conversion_rank','POE_ISFP_pct','POE_makeup_rank','POE_makeup']]
```
