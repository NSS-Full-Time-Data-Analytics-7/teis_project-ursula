```python
duplicates = compiled_monthly.duplicated(subset = ['Child ID'], keep=False)
dups_df = compiled_monthly[duplicates].sort_values(by = 'Child ID')
```

```python
duplicates = compiled_monthly.duplicated(subset = ['Child ID'], keep='first')
dups_df = compiled_monthly[duplicates].sort_values(by = 'Child ID')
```

```python
#this code takes the COUNT of duplicate child IDs (the # of rows in dups_df) and divides by the COUNT of all child IDs
#which gives you = "what % of the total number of referrals have duplicate child IDs?"
print(dups_df.shape[0])
print(compiled_monthly.shape[0])
dups_df.shape[0] / compiled_monthly.shape[0]
```

```python
#this shows you the unique children who have been referred more than once, there are 15,267
dups_df.drop_duplicates(subset = ['Child ID'], keep = 'first')

```
